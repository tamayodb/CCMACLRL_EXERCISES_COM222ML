### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x1 = [1, 2, 3, 4, 5]
x2 = [2, 4, 6, 8, 10]

plt.plot(x1)
plt.plot(x2)
plt.show()
```

```python
y1 = [2,4,6,8,10]


plt.plot(x1, y1, marker = 'o', color = "red")
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(axis= 'y')
plt.show()
```

```python
plt.plot(record_dataset['height'], record_dataset['weight'], marker = 'o', color = "red")
plt.title("Line Plot")
plt.xlabel("height")
plt.ylabel("weight")
plt.grid(axis= 'y')
plt.show()
```

```python
plt.scatter(record_dataset['height'], record_dataset['weight'], marker = 'o', color = "red")
plt.title("Scatter Plot")
plt.xlabel("height")
plt.ylabel("weight")
plt.grid()
plt.show()
```
